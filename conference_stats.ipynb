```python
play_df=pd.read_csv("/Users/benpeyton/Desktop/vandy_boot_camp/Module_20/Segment-1/Resources/play.csv")
play_df.head()
```

```python
play_df=play_df.rename(columns={"Offense Team Code": "Team Code1", "Defense Team Code": "Team Code2"})
play_df.head()
```

```python
play_df = play_df.drop(columns=["Team Code2"])
play_df
```

```python
play_df=play_df.rename(columns={"Team Code1": "Team Code"})
play_df.head()
```

```python
# Combine the data into a single dataset.
complete_stats_df = pd.merge(team_df, play_df, on=["Team Code", "Team Code"])
complete_stats_df
```

```python
complete_stats_df = complete_stats_df.sort_values(by=['Game Code', 'Play Number'])
complete_stats_df.head(50)
```

```python
# Combine the data into a single dataset.
complete_stats_df = pd.merge(complete_stats_df, conference_df, on=["Conference Code", "Conference Code"])
complete_stats_df.head(50)
```

```python
conference_stats_df = complete_stats_df.drop(columns=["Conference Code", "Name_y", "Subdivision_x"])
conference_stats_df
```

```python
conference_stats_df=conference_stats_df.rename(columns={"Name_x": "Team Name", "Name": "Conference", "Subdivision_y": "Subdivision"})
conference_stats_df
```
